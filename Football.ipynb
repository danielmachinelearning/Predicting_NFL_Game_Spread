```python
import nfl_data_py as nfl
import pandas as pd
import nflscraPy
import xgboost as xgb
```

```python
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
```

```python
past_game_stats = pd.DataFrame()

for k in years:
  season_gamelogs = nflscraPy._gamelogs(
      k
  )

  df1 = season_gamelogs[['season', 'week', 'week_day', 'event_date', 'tm_alias', 'opp_alias', 'tm_location', 'opp_location', 'tm_score', 'opp_score']]
  teams = set(df1['tm_alias'].tolist())
  for i in teams:
    df2 = df1[(df1['tm_alias'] == i) | (df1['opp_alias'] == i)]
    df2['win'] = 0
    df2['win_current'] = 0
    df2.loc[df2['tm_alias'] == i, 'win'] = 1
    df2.loc[df2['tm_alias'] == i, 'win_current'] = 1
    df3 = df2[['tm_location', 'opp_location', 'tm_score', 'opp_score', 'win']]
    df4 = df3.shift(1).fillna(0)
    df4['team'] = i
    df4.columns = ['tm_location_past', 'opp_location_past', 'tm_score_past', 'opp_score_past', 'win_past', 'team']
    df5 = df2[['season', 'week', 'win_current']]
    df6 = pd.concat([df5, df4], axis=1)
    past_game_stats = pd.concat([past_game_stats, df6])

print(past_game_stats)
```

```python
game_stats = pd.DataFrame()

for k in years:
  season_gamelogs = nflscraPy._gamelogs(
      k
  )

  df1 = season_gamelogs[['season', 'week', 'week_day', 'event_date', 'tm_alias', 'opp_alias', 'tm_location', 'opp_location', 'tm_score', 'opp_score']]
  df1['spread'] = df1['tm_score'] - df1['opp_score']
  df1.drop(['tm_score', 'opp_score'], axis=1, inplace=True)
  teams = set(df1['tm_alias'].tolist())
  for i in teams:
    df2 = df1[(df1['tm_alias'] == i) | (df1['opp_alias'] == i)]
    game_stats = pd.concat([game_stats, df2])

game_stats
```

```python
conditions = [
    game_stats['opp_location'].eq('H')
]
choices = [game_stats['opp_alias']]
game_stats['location'] = np.select(conditions, choices, default=game_stats['tm_alias'])
game_stats
```

```python
df_temp = pd.merge(past_game_stats, game_stats, left_on=['season', 'week', 'team'], right_on=['season', 'week', 'tm_alias'], how='right')
df_team_stats = pd.merge(past_game_stats, df_temp, left_on=['season', 'week', 'team'], right_on=['season', 'week', 'opp_alias'], how='right')
df_team_stats
```

```python
qb_info = pd.DataFrame()

for k in years:
  df = nfl.import_weekly_data([k])
  df2 = df[df['position'] == 'QB']
  teams = set(df2['recent_team'].tolist())
  for i in teams:
    qbs = set(df2[(df2['recent_team'] == i)]['player_name'].tolist())
    for j in qbs:
      df3 = df2[(df2['recent_team'] == i) & (df2['player_name'] == j)].sort_values(by=['week'], ascending=True,ignore_index=True).fillna(0)
      df4 = df3[['completions', 'attempts',
        'passing_yards', 'passing_tds', 'interceptions', 'sacks', 'sack_yards',
        'sack_fumbles', 'sack_fumbles_lost', 'passing_air_yards',
        'passing_yards_after_catch', 'passing_first_downs', 'passing_epa',
        'passing_2pt_conversions', 'pacr', 'dakota', 'carries', 'rushing_yards',
        'rushing_tds', 'rushing_fumbles', 'rushing_fumbles_lost',
        'rushing_first_downs', 'rushing_epa', 'rushing_2pt_conversions',
        'receptions', 'targets', 'receiving_yards', 'receiving_tds',
        'receiving_fumbles', 'receiving_fumbles_lost', 'receiving_air_yards',
        'receiving_yards_after_catch', 'receiving_first_downs', 'receiving_epa',
        'receiving_2pt_conversions', 'racr', 'target_share', 'air_yards_share',
        'wopr', 'special_teams_tds', 'fantasy_points', 'fantasy_points_ppr']]
      df5 = df4.shift(1).fillna(0)
      df6 = df3[['player_name', 'position', 'recent_team', 'season', 'week']]
      df7 = pd.concat([df6, df5], axis=1)
      qb_info = pd.concat([qb_info, df7])
      qb_info.drop(['player_name', 'position'], axis=1, inplace=True)

print(qb_info)
```

```python
df_temp = pd.merge(df_team_stats, qb_info, left_on=['season', 'week', 'tm_alias'], right_on=['season', 'week', 'recent_team'], how='right')
df_temp = df_temp.fillna(0)
df_temp = df_temp.drop_duplicates()
df_final = pd.merge(df_temp, qb_info, left_on=['season', 'week', 'opp_alias'], right_on=['season', 'week', 'recent_team'], how='right')
df_final = df_final.fillna(0)
df_final = df_final.drop_duplicates()
df_final = df_final.sort_values(['season', 'tm_alias', 'week'], ascending=[True, True, True])
df_final
```

```python
from xgboost import XGBClassifier, XGBRegressor, plot_importance
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.preprocessing import MinMaxScaler, PowerTransformer, StandardScaler, Normalizer
from sklearn.decomposition import PCA

scaler = MinMaxScaler()
#df_overall = df_overall + 1e-6

df_overall.columns = df_overall.columns.astype(str)
scaler.fit(df_overall)
input = scaler.transform(df_overall)

seed = 7
test_size = 0.1
X_train, X_test, y_train, y_test = train_test_split(np.array(input), np.array(wins), test_size=test_size, random_state=seed)
```

```python
 # fit model no training data

#clf = RandomForestRegressor(n_estimators = 2000)
clf = XGBRegressor(n_estimators=2000)

clf.fit(X_train, y_train)

# performing predictions on the test dataset
y_pred = clf.predict(X_test)

# metrics are used to find accuracy or error
from sklearn import metrics

# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL:", metrics.root_mean_squared_error(y_test, y_pred))
```

```python
from keras.layers import Dense, Activation, Dropout
from keras.models import Sequential

model = Sequential()
model.add(Dense(250, activation = 'relu', input_dim = 160))
model.add(Dropout(0.2))
model.add(Dense(units = 1000, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 250, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 1))
model.compile(optimizer = 'Nadam', loss = 'mean_squared_error')

# Fitting the ANN to the Training set
model.fit(X_train, y_train, shuffle=True, batch_size = 64, epochs = 200, validation_data=(X_test, y_test))

y_pred = model.predict(X_test)
```
